```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

# Check basic info of the dataset
print(train_data.info())

# Summary statistics for numerical columns
print("\nTraining Data Summary Statistics:")
print(train_data.describe())

# Check for missing values
print("\nMissing Values in Training Data:")
print(train_data.isnull().sum())

# Plot heatmap for missing values
plt.figure(figsize=(10, 6))
sns.heatmap(train_data.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data in Training Set")
plt.show()

# Histogram for numerical features
train_data.hist(bins=20, figsize=(10, 8))
plt.suptitle('Distribution of Numeric Features')
plt.show()

# Correlation matrix for numerical columns
numeric_cols = train_data.select_dtypes(include=['float64', 'int64']).columns  # Select numeric columns only
corr_matrix = train_data[numeric_cols].corr()  # Compute correlation matrix for numeric columns

# Plot correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Count plot for survival distribution
sns.countplot(data=train_data, x='Survived')
plt.title('Survival Count')
plt.show()

# Boxplot for Age distribution by Survival
sns.boxplot(data=train_data, x='Survived', y='Age')
plt.title('Age Distribution by Survival')
plt.show()
```
